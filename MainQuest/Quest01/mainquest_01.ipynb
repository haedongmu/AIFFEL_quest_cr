```python
import random as r

class Account:
    bank_name = "SC은행"
    account_num = 0            # 클래스 변수로 계좌 개수 관리
    customer_dict = {}         # 클래스 변수로 계좌 정보 관리
    account_history = {}       # 클래스 변수로 입출금내역 관리

    def __init__(self, name, balance):
        self.customer_dict = {}   #고객의 계좌목록
        account_number = self.get_account_number() #계좌번호생성
        self.account_number = account_number       #현재 객체의 계좌번호(입출금시 필요)
        Account.customer_dict[account_number] = {     #계좌번호를 키 값으로 딕셔너리에 계좌정보를 저장
            'bank_name': Account.bank_name,  #은행이름
            'name': name,                    #예금주
            'balance': balance,              #잔고
            'deposit_count': 1               #입금횟수
        }

        Account.account_num += 1

        #입금내역 기록
        self.update_history('입금', '입금', balance)

    #계좌번호 생성
    def get_account_number(self):
        return self.get_rand_digit(3)+"-"+self.get_rand_digit(2)+"-"+self.get_rand_digit(6)

    #랜덤숫자생성
    def get_rand_digit(self, num_digits):
        return f"{r.randint(0, 10 ** (num_digits-1)):0{num_digits}}"

    #숫자 세자리마다 콤마 넣기
    def formatted_number(self, number):
        return "{:,}".format(number)

    #일정금액 이상 계좌조회
    def display_search_account(self, search_amount):
        self.get_account_num()
        for account_number in Account.customer_dict:
            info = Account.customer_dict[account_number]
            balance = info['balance']
            if balance >= search_amount:
                self.display_info(account_number)


    #생성된 계좌의 개수 출력
    def get_account_num(self):
        print(f"생성된 계좌의 총 개수:{Account.account_num}")

    #계좌정보 출력
    def display_info(self, account_number): #은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원
        bank_name = Account.customer_dict[account_number]['bank_name']
        name = Account.customer_dict[account_number]['name']
        balance = self.formatted_number(Account.customer_dict[account_number]['balance'])
        print(f"은행이름: SC은행, 예금주: {name}, 계좌번호: {account_number}, 잔고: {balance}원")

    #입금
    def deposit(self, deposit_amount):
        #1원 이상만 입금가능
        if deposit_amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
        else:
            #입금처리
            Account.customer_dict[self.account_number]['balance'] += deposit_amount
            Account.customer_dict[self.account_number]['deposit_count'] += 1 #입금횟수 증가

            #입금내역 기록
            self.update_history('입금', '입금', deposit_amount)

            current_balance = Account.customer_dict[self.account_number]['balance']  #현재잔고 읽기
            deposit_count = Account.customer_dict[self.account_number]['deposit_count']
            if deposit_count == 5: #입금횟수가 5회가 될때마다 이자를 잔액 대비 1% 지급
                interest =  int(current_balance * 0.01) #1원 미만 절삭
                Account.customer_dict[self.account_number]['balance'] += interest
                Account.customer_dict[self.account_number]['deposit_count'] = 0 #입금횟수 초기화

                #입금내역 기록
                self.update_history('입금', '이자지급', interest)

                print(f"{interest}의 이자가 지급되었습니다.")

    #출금
    def withdraw(self, withdraw_amount):
        #잔고조회
        current_balance = Account.customer_dict[self.account_number]['balance']

        #잔고이상 출금 불가능
        if withdraw_amount > current_balance:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")
        else:
            Account.customer_dict[self.account_number]['balance'] -= withdraw_amount

            #출금내역 기록
            self.update_history('출금', '출금', -1*withdraw_amount)

     #입금내역 업데이트
    def deposit_history(self):
        self.check_history('입금')

    #출금내역 업데이트
    def withdraw_history(self):
        self.check_history('출금')

    def check_history(self, check_gubun):
        for index in Account.account_history[self.account_number]:
            gubun = Account.account_history[self.account_number][index]['gubun']
            if gubun == check_gubun:
                self.display_history(index)

    def display_history(self, index):
        title = Account.account_history[self.account_number][index]['title']
        amount = Account.account_history[self.account_number][index]['amount']
        balance = Account.account_history[self.account_number][index]['balance']
        print(f"{index+1}회: {title}\t 금액: {amount}\t 잔액: {balance}")

    def update_history(self, gubun, title, amount):
        index = 0

        #입출금내역의 인덱스
        if self.account_number in Account.account_history and isinstance(Account.account_history[self.account_number], dict):
            index = len(Account.account_history[self.account_number])

        #입출금내역 저장
        if self.account_number not in Account.account_history:
            Account.account_history[self.account_number] = {}

        Account.account_history[self.account_number][index] = {
            'gubun': gubun,
            'title': title,
            'amount': abs(amount),
            'balance': Account.customer_dict[self.account_number]['balance'] + amount
        }

#계좌 3개 생성
first_account = Account('권중', 1000000)
second_account = Account('정우성', 500000)
third_account = Account('김태희', 1500000)

#잔고 100만원 이상 고객정보 출력
first_account.display_search_account(1000000)

#입금하기
my_account = Account('권중', 10000)

for i in range(10):
    amount = r.randint(1, 10) * 1000
    my_account.deposit(amount)
my_account.deposit(-1000)
my_account.withdraw(1000000)
my_account.withdraw(1000)

#입금내역 출력
my_account.deposit_history()
print('-'*50)
#출금내역 출력
my_account.withdraw_history()
```

```python
import random as r

class Character:
    def __init__(self, name, level, hp, ap, dp):
        self.name = name
        self.level = level
        self.hp = hp
        self.ap = ap
        self.dp = dp

    #현재 체력 검사
    def is_alive(self):
        if self.hp > 0:
            return True
        else:
            return False

    #방어자의 체력 계산
    def take_damage(self, defender, damage):
        if defender.dp <= damage: #본인의 방어력이 데이미와 같거나 작다면 (받은 데미지- 본인의 방어력) 만큼 현재 체력이 감소함
            defender.hp -= (damage - defender.dp)

    #공격자가 방어자를 공격함
    def attack_target(self, attacker, defender):
        print(f"{attacker.name}이 {defender.name}에게 {attacker.ap}만큼 공격했다...!")
        damage = r.randint(1, attacker.ap)  #데미지는 1부터 공격력 사이의 랜덤 정수
        defender.take_damage(defender, damage)

class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.xp = 0 #경험치

    #경험치 얻기
    def gein_experience(self, monster_level):
        self.xp += (monster_level * 20)

    def level_up(self):
        if self.xp >= 50:
            self.ap += 10
            self.dp += 5

class Monster(Character):
    def __init__(self, name, level):
        self.name = name
        self.level = level
        self.hp = r.randint(10, 30) * level
        self.ap = r.randint(5, 20) * level
        self.dp = r.randint(1, 5) * level

def battle(player, monster_dict):
    for monster_name in monster_dict:
        if player.is_alive():
            print(f'{monster_name} 과의 전투를 시작합니다.')
            monster = Monster(monster_name, monster_dict[monster_name])
            while player.is_alive() and monster.hp > 0: #플레이어와 몬스터가 모두 생존한 경우 전투 계속함
                #플레이어 공격
                player.attack_target(player, monster)

                #몬스터 공격
                monster.attack_target(monster, player)

            if player.is_alive():
                player.gein_experience(monster.level)
                player.level_up()
                print("전투 승리!")
                print(f"{monster_name} 을 이겼다!")
            else:
                print("전투 패배..")
                print("게임오버")

    print("모든 전투를 마치고 게임을 종료합니다.")

def main():
    monster_dict = {
        '슬라임': 1,
        '고블린': 2,
        '오크': 3
    }

    player_name = input("플레이어의 이름을 입력해 주세요..")
    player = Player(player_name)

    #전투시작
    battle(player, monster_dict)

main()

```
