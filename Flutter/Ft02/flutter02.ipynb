```python
import 'package:flutter/material.dart';

void main() {
  runApp(MyApp());
}

class MyApp extends StatelessWidget {
  @override
  Widget build(BuildContext context) {
    return MaterialApp(
        debugShowCheckedModeBanner: false,
        home: Scaffold(
        appBar: AppBar(
          backgroundColor: Colors.blue,
          // AppBar의 배경색을 파란색으로 설정(테마에 따라 달라지기 때문에 명시적으로 지정해야 함)
          centerTitle: true,
          // 제목을 중앙에 배치
          title: TopScreen(),
        ),
        body: BodyScreen()
      )
    );
  }
}

class TopScreen extends StatelessWidget {
  @override
  Widget build(BuildContext context) {
    return Row(
        children: [
          Image.asset(
            'images/top.png',
            width: 50,
          ),
          SizedBox(width: 10),
          Container(
            width: 250,
            alignment: Alignment.center,
            child: Text(
                textAlign: TextAlign.center,
                '플러터 앱 만들기'
            ),
          )
        ]
    );
  }
}

class BodyScreen extends StatelessWidget {
  @override
  Widget build(BuildContext context) {
    final List<Color> colors = [
      Colors.red,
      Colors.orange,
      Colors.yellow,
      Colors.green,
      Colors.blue,
    ];

    final List<double> sizes = [300, 240, 180, 120, 60];

    return Center(
        child: Column(
          mainAxisAlignment: MainAxisAlignment.center, // 세로축에서 중앙 정렬
          children: [
            ElevatedButton(
                onPressed: (){
                  print('버튼이 눌렸습니다.');
                },
                style: ElevatedButton.styleFrom(
                  backgroundColor: Colors.purple, // 버튼 배경색 보라색으로 설정
                  foregroundColor: Colors.white, // 버튼 글씨색 흰색
                ),
                child: Text('Text')
            ),
            SizedBox(height: 30),
            Stack(
              children: List.generate(
                colors.length,
                (index) => Container(
                  width: sizes[index],
                  height: sizes[index],
                  color: colors[index],
                ),
              ),
            ),
          ],
        ),
    );
  }
}
```
